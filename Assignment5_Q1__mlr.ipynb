```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.regressionplots import influence_plot
import statsmodels.formula.api as smf
import numpy as np
```

```python
data=pd.read_csv("50_Startups.csv")
data
```

```python
#EDA
data.info()
```

```python
data.isna().sum()
```

```python
data=data.rename({'R&D Spend':'RDS','Administration':'Admini','Marketing Spend':'MarketS'},axis=1)
data
```

```python
data.describe()
```

```python
#Scatterplot between variables along with histograms
#Format the plot background and scatter plots for all the variables
sns.set_style(style='darkgrid')
sns.pairplot(data)
```

```python
#Model Building
model=smf.ols('Profit~MarketS+RDS+Admini+RDS',data=data).fit()
```

```python
#model testing
model.params
```

```python
# Finding tvalues and pvalues
print(model.tvalues, '\n', model.pvalues)
```

```python
#R squared values
(model.rsquared,model.rsquared_adj)
```

```python
#Simple Linear Regression Model
slr_a=smf.ols("Profit~Admini",data=data).fit()
slr_a.tvalues , slr_a.pvalues  # ADMS has in-significant pvalue
slr_a.summary()
```

```python
slr_m=smf.ols("Profit~MarketS",data=data).fit()
slr_m.tvalues , slr_m.pvalues  # MKTS has significant pvalue
slr_m.summary()
```

```python
mlr_am=smf.ols("Profit~Admini+MarketS",data=data).fit()
mlr_am.tvalues , mlr_am.pvalues  # varaibles have significant pvalues
mlr_am.summary()
```

```python
#Model Validation Techniques
#Two Techniques: 1. Collinearity Check
# 1) Collinearity Problem Check
# Calculate VIF = 1/(1-Rsquare) for all independent variables

rsq_r=smf.ols("RDS~Admini+MarketS",data=data).fit().rsquared
vif_r=1/(1-rsq_r)

rsq_a=smf.ols("Admini~RDS+MarketS",data=data).fit().rsquared
vif_a=1/(1-rsq_a)

rsq_m=smf.ols("MarketS~RDS+Admini",data=data).fit().rsquared
vif_m=1/(1-rsq_m)

# Putting the values in Dataframe format
d1={'Variables':['RDS','Admini','MarketS'],'Vif':[vif_r,vif_a,vif_m]}
Vif_df=pd.DataFrame(d1)
Vif_df
```

```python
#Residual Analysis
#Test for Normality of Residuals (Q-Q Plot)
import statsmodels.api as sm
qqplot=sm.qqplot(model.resid,line='q') # line = 45 to draw the diagnoal line
plt.title("Normal Q-Q plot of residuals")
plt.show()
```

```python
list(np.where(model.resid<-20000))  # outliar detection from above QQ plot of residuals
```

```python
#Test for Homoscedasticity or Heteroscedasticity
def standard_values( vals ):
    return (vals - vals.mean())/vals.std()
```

```python
plt.scatter(standard_values(model.fittedvalues),standard_values(model.resid))
plt.title('Residual Plot')
plt.xlabel('standardized fitted values')
plt.ylabel('standardized residual values')
plt.show() 
```

```python
#Residual Vs Regressors
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "RDS", fig=fig)
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "Admini", fig=fig)
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "MarketS", fig=fig)
plt.show()
```

```python
#Model Deletion Diagnostics (checking Outliers or Influencers)
#Two Techniques : 1. Cook's Distance
model_influence = model.get_influence()
(c, _) = model_influence.cooks_distance
```

```python
#Plot the influencers values using stem plot
fig = plt.subplots(figsize=(20, 7))
plt.stem(np.arange(len(data)), np.round(c, 3))
plt.xlabel('Row index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
#index and value of influencer where c is more than .5
(np.argmax(c),np.max(c))
```

```python
# 2. Leverage Value using High Influence Points : Points beyond Leverage_cutoff value are influencers
from statsmodels.graphics.regressionplots import influence_plot
influence_plot(model)
plt.show()
```
